#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
app_credit.info()
````

````python
app_skl = app_credit.copy()
````

````python
app_skl = pd.get_dummies(app_skl, columns=['CODE_GENDER'], prefix='gen', prefix_sep='_')

app_skl = pd.get_dummies(app_skl, columns=['FLAG_OWN_CAR'], prefix='car', prefix_sep='_')

app_skl = pd.get_dummies(app_skl, columns=['FLAG_OWN_REALTY'], prefix='rlt', prefix_sep='_')

app_skl = pd.get_dummies(app_skl, columns=['NAME_INCOME_TYPE'], prefix='inc', prefix_sep='_')

app_skl = pd.get_dummies(app_skl, columns=['NAME_EDUCATION_TYPE'], prefix='edu', prefix_sep='_')

app_skl = pd.get_dummies(app_skl, columns=['NAME_FAMILY_STATUS'], prefix='fam', prefix_sep='_')

app_skl = pd.get_dummies(app_skl, columns=['NAME_HOUSING_TYPE'], prefix='hou', prefix_sep='_')

app_skl = pd.get_dummies(app_skl, columns=['OCCUPATION_TYPE'], prefix='occ', prefix_sep='_')

app_skl.info()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
default_1 = app_credit[app_credit['interesse'] == 1]

var_pod = pd.crosstab(default_1['FLAG_EMAIL'],
                      default_1['FLAG_OWN_CAR'])
var_pod
````

Neste contexto dos dados apresentados, a variável 'FLAG_EMAIL' pode ser considerada mais preditiva, pois apresenta uma maior variação entre as categorias, indicando uma relação mais robusta com o desfecho de interesse.
